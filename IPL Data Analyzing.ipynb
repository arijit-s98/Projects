### Which team won by batting first and by how much runs

```python
batting_first = df1[df1['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
batting_first.loc[batting_first['win_by_runs'].idxmax()]
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.xlabel('runs')
plt.ylabel('number of matches')
plt.show()
```

### Which team had won by maximum wickets

```python
df1.loc[df1['win_by_wickets'].idxmax()]
```

```python
batting_second = df1[df1['win_by_wickets']!=0]
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.hist(df1['win_by_wickets'])
plt.xlabel('wickets')
plt.ylabel('number of wickets')
plt.show()
```

### Plotting a pie chart on winning value on batting first

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

### plot winning value on batting second

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
